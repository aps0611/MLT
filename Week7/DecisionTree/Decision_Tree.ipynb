```python
## Decision Tree

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
## load the iris dataset

df = sns.load_dataset('iris')
df.head(5)
```

```python
df['species'] = df['species'].map({'setosa' : 0,
                                   'versicolor' : 1,
                                   'virginica' : 2
                                   })
```

```python
## use sklearn for Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy')
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])
```

```python
## Constructing the decision tree

from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(classifier, filled = True)
```

```python
## Use post-pruning when Dataset is small
## Use pre-pruning when Dataset is large

## Read more on CART and Pre-proning and Post-pruning and grid first search
```

```python

from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', max_depth = 2)
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(classifier, filled = True)
```

```python
## using gini
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'gini')
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(classifier, filled = True)
```

```python
### read more on gridsearch CV and crossval score
### Decision Tree regressor
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

params = {'max_depth' : [2,4,6,8,10,12],
          'min_samples_split' : [2,3,4],
          'min_samples_leaf': [1,2]}
clf = tree.DecisionTreeClassifier()
gcv = GridSearchCV(estimator = clf, param_grid= params)
gcv.fit(X_train, y_train)
```
